```python
# PART 1

# downloading packages
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import statsmodels.api as sm

# downloading clean dataset
# importing the dataset
clean_smoking = pd.read_csv("C:/Users/justi/Introduction-to-Python-S1-M1/Python-Introduction-Smoking-Prevalence/data/clean/clean_smoking_df.csv")
clean_educ =pd.read_csv("C:/Users/justi/Introduction-to-Python-S1-M1/Python-Introduction-Smoking-Prevalence/data/clean/clean_educ_df.csv")
clean_smoking.head()
clean_educ.head()
```

```python
# PART 2: Merging both dataset

merged_df = pd.merge(clean_smoking, clean_educ, how='left', on=['geo', 'sex'])
merged_df.head()
```

```python
# rename / shorten some names
renames = { "obs_value_x": "value_smoking",
           "obs_value_y":"value_educ_leavers"}

merged_df = merged_df.rename(columns = renames)
merged_df.head()
```

```python
# PART 2: mutivariate regression

# preparing data

merged_df.head()
merged_df.isnull().sum()
merged_df.head()
```

```python
# subset sex variable

merged_df['sex_binary'] = merged_df['sex'].map({'Females': 0,
                                                        'Males': 1})
merged_df['sex_binary'] = pd.to_numeric(merged_df['sex_binary'],errors='coerce')
merged_df['value_educ_leavers'] = pd.to_numeric(merged_df['value_educ_leavers'],errors='coerce')
merged_df = merged_df.dropna(subset=['value_educ_leavers'])

merged_df.head()
```

```python
# set the variables

X_multiple = merged_df[['sex_binary', 'value_educ_leavers']]  # independent variable(s)
y = merged_df['value_smoking']   # dependent variable
print(X_multiple)
```

```python
# Create and fit the model

model_multiple = LinearRegression()
model_multiple.fit(X_multiple, y)
```

```python
# Print the results

print("Intercept (β₀):", model_multiple.intercept_)
print("Coefficients(β₁+ β2)):", model_multiple.coef_)
```

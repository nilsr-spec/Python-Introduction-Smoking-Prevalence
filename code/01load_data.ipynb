```python
# PART 3

# cleaning the names
df1.columns =(df1.columns.str.lower().str.strip().str.replace(" ", "_"))
print(df1.columns)

df2.columns =(df2.columns.str.lower().str.strip().str.replace(" ", "_"))
print(df2.columns)
```

```python
# correct variable types

df1["last_update"] = pd.to_datetime(df1["last_update"], errors="coerce")
df2["last_update"] = pd.to_datetime(df2["last_update"], errors="coerce")

categorial_cols = ["freq", "unit", "age", "sex", "geo"]
df1[categorial_cols] = df1[categorial_cols].astype("category")
df2[categorial_cols] = df2[categorial_cols].astype("category")

df1.head()
df2.head()
```

```python
# drop useless and NaN full columns
df1 = df1.drop(columns=["conf_status", "obs_flag", "dataflow"])
df2 = df2.drop(columns=["conf_status", "obs_flag", "dataflow"])

df1.head()
df2.head()
```

```python
# PART 3

# keep only useful columns

useful_cols = ["STRUCTURE_NAME", "Age class", "Sex", "Unit of measure", "Geopolitical entity (reporting)", "TIME_PERIOD", "OBS_VALUE"]

educ_df = educ_df[useful_cols] 
smoking_df = smoking_df[useful_cols]

educ_df.head()
smoking_df.head()
```

```python
# cleaning the names
educ_df.columns =(educ_df.columns.str.lower().str.strip().str.replace(" ", "_"))

smoking_df.columns =(smoking_df.columns.str.lower().str.strip().str.replace(" ", "_"))

# rename / shorten some names
renames = {
    "age_class": "age",
    "geopolitical_entity_(reporting)": "geo"
}

educ_df = educ_df.rename(columns=renames)
smoking_df = smoking_df.rename(columns=renames)

print(educ_df.columns)
print(smoking_df.columns)
```

```python
# correct variable types

categorial_cols = ["age", "sex", "geo"]
educ_df[categorial_cols] = educ_df[categorial_cols].astype("category")
smoking_df[categorial_cols] = smoking_df[categorial_cols].astype("category")

educ_df.head()
smoking_df.head()
```

```python
# PART 4

# returning a clean version of the dataset (Justine)

educ_df.to_csv("C:/Users/justi/Introduction-to-Python-S1-M1/Python-Introduction-Smoking-Prevalence/data/clean_educ_df.csv")
smoking_df.to_csv("C:/Users/justi/Introduction-to-Python-S1-M1/Python-Introduction-Smoking-Prevalence/data/clean_smoking_df.csv")
```
